```python
#library import
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.cross_validation import train_test_split

#data reading
train = pd.read_csv('numerai_training_data.csv')
target = train.target
train = train.drop('target', axis = 1)
test = pd.read_csv('numerai_tournament_data.csv')
t_id = test.t_id
test = test.drop('t_id', axis = 1)


#cross_validation
X_train, X_test, target_train, target_test = train_test_split(train, target, random_state=1234, test_size=0.3)

clf = xgb.XGBClassifier(objective="binary:logistic",
                max_depth = 5,
                n_estimators=1000,
                learning_rate=0.03, 
                nthread=4,
                subsample=0.75,
                colsample_bytree=0.75,
                min_child_weight = 3,
                reg_alpha=0.03,
                seed=1301)
clf.fit(X_train, target_train, early_stopping_rounds=100, eval_metric="logloss", verbose = 1,
        eval_set=[(X_train, target_train), (X_test, target_test)])

pred = clf.predict_proba(test, ntree_limit=clf.best_iteration)[:, 1]
pd.DataFrame({"t_id": t_id, "probability": pred}).to_csv('numerai_1.csv',index=False)
```

```python
ts = pd.Series(clf.booster().get_fscore())
```
